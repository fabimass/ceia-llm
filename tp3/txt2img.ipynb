```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
def print_image(image):
    plt.imshow(image)
    plt.axis("off")
    plt.show()
```

```python
# Prompt para generar la imagen
prompt = "Una persona con lentes de sol bebiendo cafe"

# Generamos la imagen
image = pipe(prompt).images[0]

# Mostramos la imagen generada
print_image(image)
```

```python
prompts = [
    "Una mariposa amarilla posando sobre un arbusto rojo",
    "Un robot jugando al futbol",
    "Un perro corriendo en la playa"
]

# Generar y mostrar imágenes
for prompt in prompts:
    image = pipe(prompt).images[0]
    print_image(image)
```
